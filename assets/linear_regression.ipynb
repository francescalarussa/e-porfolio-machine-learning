```python
years = [str(y) for y in range (2001, 2020)]
global_population=global_population[years]
```

```python
global_population.isna().sum()
```

```python
global_population = global_population.apply(pd.to_numeric, errors='coerce')
```

```python
global_population= global_population.dropna(how="all")
global_population=global_population.dropna(thresh=5)
global_population.interpolate(axis=1).bfill(axis=1).ffill(axis=1)
```

```python
global_population.isna().sum()
```

```python
I calculate the GDP and Population means and I create a new dataframe with the variable aggregate per country.
```

```python
population_mean = global_population.mean(axis=1)
```

```python
gdp_mean = global_gdp.mean(axis=1)
```

```python
gdp_per_capita = global_gdp/global_population
```

```python
gdp_per_capita = gdp_per_capita.interpolate(axis=1).bfill(axis=1).ffill(axis=1)
```

```python
gdp_pc_mean = gdp_per_capita.mean(axis=1)
```

```python
gdp_population = pd.concat([population_mean,gdp_pc_mean], axis=1)
gdp_population.columns=["population","gdp_per_capita"]
gdp_population = gdp_population.dropna()
```

```python
Calculate Correlation
```

```python
gdp_population.corr(numeric_only=True)
```

```python
The correlation between GDP and Population is weak. However, I try to run a linear regression
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error
import matplotlib.pyplot as plt
```

```python
X = gdp_population[["population"]]
Y = gdp_population[["gdp_per_capita"]]
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 0.2, random_state=42)
model= LinearRegression()
model.fit(X_train, Y_train)
Y_pred = model.predict(X_test)
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
print("MAE:", mean_absolute_error(Y_test,Y_pred))
print("R2:", r2_score(Y_test, Y_pred))
```

```python
Create a Scatter Plot
```

```python
plt.scatter(X_train, Y_train, color = "blue")
plt.plot(X_train, model.predict(X_train), color="red")
plt.xlabel("Population")
plt.ylabel("GDP")
plt.show()
```
